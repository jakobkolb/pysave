```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
taus = np.logspace(0,2,100)
plt.rc('font', size=24)
fig, (ax1) = plt.subplots(figsize=(15,10))


ax1.semilogx(taus, 0.5*0.8/(np.exp(taus*0.8)-1),'g',lw=3,label=r'$\delta=0.80$')

ax1.semilogx(taus, 0.2*0.5/(np.exp(taus*0.2)-1),'blue',lw=3,label=r'$\delta=0.20$')
ax1.semilogx(taus, 0.06*0.5/(np.exp(taus*0.06)-1),'orange',lw=3,label=r'$\delta=0.06$')
ax1.semilogx(taus, 0.01*0.5/(np.exp(taus*0.01)-1),'r',lw=3,label=r'$\delta=0.01$')

ax1.legend(loc='upper right')
ax1.set_ylabel(r'$\rho(\tau)$')
ax1.set_xlabel(r'$\tau$')
#ax1.set_yscale('log')
#fig.savefig('rhotau.pdf')
```
